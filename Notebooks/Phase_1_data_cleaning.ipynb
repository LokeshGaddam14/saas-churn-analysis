```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-whitegrid')
print("Libraries loaded successfully")
```

```python
df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')

print("Shape:", df.shape)
print("\nFirst 5 rows:")
df.head()
```

```python
print("=== Dataset Info ===")
df.info()

print("\n=== Column names ===")
print(df.columns.tolist())
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

print("TotalCharges dtype now:", df['TotalCharges'].dtype)
print("Null values created:", df['TotalCharges'].isnull().sum())
```

```python
print("Missing values before:")
print(df.isnull().sum()[df.isnull().sum() > 0])

df['TotalCharges'].fillna(0, inplace=True)

print("\nMissing values after:")
print(df.isnull().sum().sum(), "nulls remaining")
```

```python
df['Churn_Binary'] = df['Churn'].map({'Yes': 1, 'No': 0})

print("Churn distribution:")
print(df['Churn'].value_counts())
print("\nChurn rate:", round(df['Churn_Binary'].mean() * 100, 2), "%")
```

```python
df['SeniorCitizen'] = df['SeniorCitizen'].map({0: 'No', 1: 'Yes'})
print(df['SeniorCitizen'].value_counts())
```

```python
print("=== Numeric columns summary ===")
df.describe().round(2)
```

```python
dupes = df.duplicated().sum()
print(f"Duplicate rows: {dupes}")

df.drop_duplicates(inplace=True)
print(f"Final dataset shape: {df.shape}")
```

```python
import os

# Create the 'data' directory if it doesn't exist
os.makedirs('data', exist_ok=True)

df.to_csv('data/telco_churn_cleaned.csv', index=False)
print("Cleaned dataset saved!")
print(f"Rows: {df.shape[0]} | Columns: {df.shape[1]}")
```
